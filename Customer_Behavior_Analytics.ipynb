```python
# Missing values in gender were filled with Not Specified to avoid incorrect assumptions

df_clean['gender'] = df_clean['gender'].fillna('Not Specified')
```

```python
# Device type varies between new and returning users, so group-based imputation was used

df_clean['device_type'] = df_clean.groupby('returning_user')['device_type'] \
    .transform(lambda x: x.fillna(x.mode()[0]))
```

```python
df_clean.isnull().sum()
```

```python
df_clean['time_on_site_hours'] = (df_clean['time_on_site'] / 60).round(2)
```

```python
df_clean['user_id'] = df_clean['user_id'].astype(int)
df_clean['age'] = df_clean['age'].astype(int)
df_clean['pages_viewed'] = df_clean['pages_viewed'].astype(int)
df_clean['previous_purchases'] = df_clean['previous_purchases'].astype(int)
df_clean['cart_items'] = df_clean['cart_items'].astype(int)
df_clean['discount_seen'] = df_clean['discount_seen'].astype(int)
df_clean['ad_clicked'] = df_clean['ad_clicked'].astype(int)
df_clean['returning_user'] = df_clean['returning_user'].astype(int)
df_clean['purchase'] = df_clean['purchase'].astype(int)

```

```python
df_clean['time_on_site'] = df_clean['time_on_site'].round(2)

df_clean['avg_session_time'] = df_clean['avg_session_time'].round(2)
df_clean['bounce_rate'] = df_clean['bounce_rate'].round(2)
df_clean['time_on_site_hours'] = df_clean['time_on_site_hours'].round(2)
```

```python
df_clean['gender'] = df_clean['gender'].str.title()

df_clean['device_type'] = df_clean['device_type'].str.title()
```

```python
df_clean['age_group'] = pd.cut(

    df_clean['age'],

    bins=[18, 25, 35, 45, 59],

    labels=[
        'Young Adults',
        'Adults',
        'Mid-Age Users',
        'Senior Users'
    ],

    include_lowest=True
)
```

```python
# Created engagement_score using scaled values of time, pages, and session time
'''Normalization was applied to bring all behavioral features

into a common scale between 0 and 1, ensuring balanced

contribution to the engagement score calculation.'''

df_clean['engagement_score'] = (

    0.5 * (df_clean['cart_items'] / df_clean['cart_items'].max()) +

    0.3 * (df_clean['pages_viewed'] / df_clean['pages_viewed'].max()) +

    0.2 * (df_clean['time_on_site'] / df_clean['time_on_site'].max())

).round(3)
```

```python
df_clean['engagement_level'] = pd.cut(
    df_clean['engagement_score'],
    bins=[0, 0.365, 0.648, 1.0],
    labels=['Low', 'Medium', 'High'],
    include_lowest=True
)
```

```python
# Created behavioral funnel stages based on number of pages viewed

# To represent customer engagement progression during website browsing

def funnel_stage(row):

    if row['pages_viewed'] <= 4:

        return 'Visitor'

    elif row['pages_viewed'] <= 9:

        return 'Product Explorer'

    elif row['pages_viewed'] <= 14:

        return 'Active Browser'

    else:

        return 'Highly Interested User'

df_clean['funnel_stage'] = df_clean.apply(funnel_stage, axis=1)
```

```python
stage_counts = df_clean['funnel_stage'].value_counts()

print(stage_counts)
```

```python
stage_percentage = (

    df_clean['funnel_stage']

    .value_counts(normalize=True) * 100

)

print(stage_percentage.round(2))
```

```python
# Created customer behavior segments based on retention,

# engagement, purchase intent, and browsing activity

def customer_stage(row):

     # Users with high bounce rate are considered low retention users
    if row['bounce_rate'] > 70:
        return 'Low Retention Users'

     # Highly engaged users based on engagement score category
    elif row['engagement_level'] == 'High':
        return 'Highly Engaged Users'

    # Users showing strong purchase intent through cart activity
    elif row['cart_items'] >= 4:
        return 'High Intent Users'

   # Users actively exploring products through page views
    elif row['pages_viewed'] > 10:
        return 'Interested Users'

    else:
        # Users with lower interaction levels
        return 'Casual Visitors'

df_clean['customer_stage'] = df_clean.apply(customer_stage, axis=1)


```

```python
df_clean['customer_stage'].value_counts()
```

```python
df_clean.head(10)
```

```python
# Suggested actions based on user behavior
def recommend(row):

    if row['customer_stage'] == 'High Intent Users':

        return 'Offer Fast Checkout Discount'

    elif row['customer_stage'] == 'Highly Engaged Users':

        return 'Recommend Premium Products'

    elif row['customer_stage'] == 'Active Browsers':

        return 'Show Trending Products'

    elif row['customer_stage'] == 'Low Retention Users':

        return 'Send Personalized Offers'

    else:

        return 'Recommend Popular Products'

df_clean['customer_recommendation'] = df_clean.apply(recommend, axis=1)

```

```python
df_clean.head(10)
```

```python
df_clean['predicted_buyer'] = (

    (df_clean['engagement_level'] == 'High') |

    (

        (df_clean['engagement_level'] == 'Medium') &

        (df_clean['cart_items'] > 2)

    ) |

    (

        (df_clean['engagement_level'] == 'Low') &

        (df_clean['cart_items'] > 4)

    )

).astype(int)


df_clean['buyer_label'] = df_clean['predicted_buyer'].map({
    1: 'Buyer',
    0: 'Non-Buyer'
})
```
